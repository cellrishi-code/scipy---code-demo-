```python
# let consider the equation
#  2x + 3y = 5
# 5x + 2 y = 6
```

```python
## now main point  #method-->1

a = np.array([[2,3],[5, 2]])

b = np.array([5,6])

result = np.linalg.inv(a).dot(b)
print(result)
```

```python
## method ---> 2
a = np.array([[2,3],[5, 2]])

b = np.array([5,6])

result = np.linalg.solve(a,b)
print(result)
```

```python
## method 3

from scipy.linalg import solve
a = np.array([[2,3],[5, 2]])

b = np.array([5,6])

result = solve(a,b)
print(result)

```

```python
from scipy.linalg import inv    ## method 4
import numpy as np

A = np.array([[2, 1],
              [1, 3]])

B = np.array([5, 6])

X = inv(A).dot(B)

print(X)
```
