```python
from scipy import constants as co
import numpy as np
```

```python
print(co.pi) # pie
```

```python
## integration in scipy
from scipy import integrate as int

result , error = int.quad(lambda x : x**5 , 0 , 10)
```

```python
print(result)
```

```python
## linear algrbra
from scipy import linalg as la

a = np.array([[1,2], [8,8]])
inverase = la.inv(a)
print(inverase)
```

```python
# stats module
from scipy import stats

a = np.array([13,82,26,18])
print(stats.norm.mean(a))
print(stats.variation(a))
```
